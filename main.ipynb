```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2019 = pd.read_csv("Zaehlstelle_Herose_2019_stuendlich_Wetter_4.csv",sep=";")
df2020 = pd.read_csv("Zaehlstelle_Herose_2020_stuendlich_Wetter_4.csv",sep=";")
df2021 = pd.read_csv("Zaehlstelle_Herose_2021_stuendlich_Wetter_5.csv",sep=";")
df2022 = pd.read_csv("Zaehlstelle_Herose_2022_stuendlich_Wetter_3.csv",sep=";")
df2023 = pd.read_csv("Zaehlstelle_Heros√©_2023_stuendlich_Wetter.csv", sep=";")
years = [df2019,df2020,df2021,df2022,df2023]
```

```python
df2019['Zeit'] = pd.to_datetime(df2019['Zeit'], format='%d.%m.%Y %H:%M')
df2020['Zeit'] = pd.to_datetime(df2020['Zeit'], format='%d.%m.%Y %H:%M')
df2021['Zeit'] = pd.to_datetime(df2021['Zeit'], format='%d.%m.%Y %H:%M')
df2022['Zeit'] = pd.to_datetime(df2022['Zeit'], format='%d.%m.%Y %H:%M')
df2023['Zeit'] = pd.to_datetime(df2023['Zeit'], format='%d.%m.%Y %H:%M')
```

```python
weekly_mean19 = df2019.resample('W', on='Zeit')['Fahrradbruecke'].mean()
weekly_mean23 = df2023.resample('W', on='Zeit')['Fahrradbruecke'].mean()
weekly_mean20 = df2020.resample('W', on='Zeit')['Fahrradbruecke'].mean()
weekly_mean21 = df2021.resample('W', on='Zeit')['Fahrradbruecke'].mean()
weekly_mean22 = df2022.resample('W', on='Zeit')['Fahrradbruecke'].mean()



```

```python
tag_avg23 = df2023.groupby(df2023['Zeit'].dt.date)['Fahrradbruecke'].mean()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(weekly_mean19.index, weekly_mean19.values, marker='o', linestyle='-')
plt.plot(weekly_mean20.index, weekly_mean20.values, marker='o', linestyle='-')
plt.plot(weekly_mean21.index, weekly_mean21.values, marker='o', linestyle='-')
plt.plot(weekly_mean22.index, weekly_mean22.values, marker='o', linestyle='-')
plt.plot(weekly_mean23.index, weekly_mean23.values, marker='o', linestyle='-')
plt.title('Durchschnittliche Anzahl von Fahrradfahrern pro Woche an der Fahrradbr√ºcke')
plt.xlabel('Datum')
plt.ylabel('Durchschnittliche Anzahl von Fahrradfahrern')
plt.xticks(rotation=45)  # Rotiere x-Achsenbeschriftungen um 45 Grad, um sie lesbar zu machen
plt.grid(True)
plt.tight_layout()
plt.show()
```
